## **🔹 8. Broadcasting in NumPy**
**Broadcasting** allows NumPy to **perform operations between arrays of different shapes** without explicitly reshaping them.

### ✅ **Example: Broadcasting a Scalar to an Array**
```python
arr = np.array([1, 2, 3, 4, 5])
result = arr + 10  # Adds 10 to each element
print("Original Array:", arr)
print("After Broadcasting:", result)
```
✅ **Output:**
```
Original Array: [1 2 3 4 5]
After Broadcasting: [11 12 13 14 15]
```
✔ NumPy **automatically applies the operation** without needing explicit loops.

---

### ✅ **Example: Broadcasting a 1D Array to a 2D Array**
```python
A = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2,3)
B = np.array([10, 20, 30])  # Shape (3,)

result = A + B  # B is broadcasted across rows
print("Matrix A:\n", A)
print("Vector B:\n", B)
print("Broadcasted Result:\n", result)
```
✅ **Output:**
```
Matrix A:
 [[1 2 3]
  [4 5 6]]

Vector B:
 [10 20 30]

Broadcasted Result:
 [[11 22 33]
  [14 25 36]]
```
✔ NumPy **automatically expands B** across all rows of A.

---

### ✅ **Broadcasting Rule**
For **broadcasting to work**, NumPy **compares shapes**:
1. If dimensions are **equal**, they align.
2. If one dimension is **1**, it **expands** to match the other.
3. Otherwise, **NumPy raises an error**.

✅ **Valid Example:**
```python
X = np.array([[1], [2], [3]])  # Shape (3,1)
Y = np.array([10, 20, 30])  # Shape (3,)

print(X + Y)  # Y expands to match X
```

❌ **Invalid Example (Mismatched Shapes)**:
```python
A = np.array([1, 2, 3])  # Shape (3,)
B = np.array([[10, 20], [30, 40]])  # Shape (2,2)

print(A + B)  # ERROR: Shapes (3,) and (2,2) are incompatible
```

## **🔹 10. Optimizing Performance with NumPy**
NumPy is **much faster** than Python lists because it uses **vectorized operations** and **efficient memory management**.

### ✅ **Speed Test: NumPy vs Python Lists**
```python
import time

size = 1000000

# Python List
py_list = list(range(size))
start = time.time()
py_result = [x * 2 for x in py_list]  # Multiply each element by 2
end = time.time()
print("Python List Time:", end - start, "seconds")

# NumPy Array
np_array = np.arange(size)
start = time.time()
np_result = np_array * 2  # Vectorized operation
end = time.time()
print("NumPy Time:", end - start, "seconds")
```
✅ **Output (Example Timing):**
```
Python List Time: 0.15 seconds
NumPy Time: 0.005 seconds
```
✔ **NumPy is 10-100x faster** than Python lists.

---

### 🔹 **What is Time-Series Data?**

Time-series data refers to a **sequence of observations recorded over time**, usually at **regular intervals** (daily, monthly, yearly, etc.).  
- The key characteristic is that **each data point is associated with a timestamp**.  
- The **order of data matters** because the sequence in time reveals patterns, trends, and behaviors.

🧠 **Key Features:**
- Time is the **independent variable**.
- Used to **observe, model, and forecast** behavior over time.
- Can be **univariate** (one variable over time) or **multivariate** (multiple variables tracked together).

✅ **Simple Example of Univariate Time-Series:**
```text
Date        | Stock Price (PKR)
------------|-------------------
2024-01-01  | 100
2024-01-02  | 105
2024-01-03  | 98
2024-01-04  | 110
```

✅ **Multivariate Time-Series:**
```text
Date        | Open | Close | Volume
------------|------|-------|--------
2024-01-01  | 98   | 100   | 2000
2024-01-02  | 100  | 105   | 3000
```

---

### 🔹 **Why Time-Series Data is Different from Other Data?**

| Feature             | Time-Series Data                      | Regular Tabular Data             |
|--------------------|----------------------------------------|----------------------------------|
| Order of Data      | **Very important**                     | Usually not important            |
| Index              | **Datetime or timestamp**              | Usually numerical or categorical |
| Applications       | Forecasting, trend analysis, anomaly detection | Classification, regression     |
| Visual Patterns    | Trend, seasonality, cyclic patterns    | Less likely to be time-related   |

---

### 🔹 **Key Components of Time-Series**

Understanding these helps **decompose** and **model** time-series data better:

| Component   | Description                                                                 | Example |
|-------------|-----------------------------------------------------------------------------|---------|
| **Trend**     | Long-term **increase or decrease** in the data over time.                  | 📈 Gradual rise in housing prices |
| **Seasonality** | **Regular patterns** at fixed intervals (daily, weekly, monthly, yearly)  | 🎄 Sales increase in December |
| **Cyclic**    | **Non-fixed repeating patterns** usually driven by the economy or market   | 💹 Recession and recovery cycles |
| **Noise**     | **Irregular random variations** that do not follow a pattern               | 📉 Sudden drop due to company news |

✅ **Visualization Suggestion:**  
- Trend → Smooth curve  
- Seasonality → Repeating waves  
- Cycles → Long irregular waves  
- Noise → Spikes and dips

## 🟢 **2. Working with Date-Time in Pandas**

This section focuses on how to work with **dates and times** using the **Pandas library** – an essential skill for financial and time-series analysis.

---

## 🔹 **Why Use Date-Time Features in Time-Series?**
In time-series data:
- Each observation is tied to a **specific time or date**.
- You need to **parse**, **index**, and **filter** using dates.
- Time helps us **resample**, **slice**, **visualize trends**, and perform **forecasting**.

---

## ✅ **2.1 Converting Strings to DateTime with `pd.to_datetime()`**

You may have a column like this:

```python
import pandas as pd

dates = ["2024-01-01", "2024-01-02", "2024-01-03"]
df = pd.DataFrame({"Date": dates, "Price": [100, 105, 103]})

df["Date"] = pd.to_datetime(df["Date"])  # Convert string to datetime
print(df.dtypes)
```

✅ **Output:**
```
Date     datetime64[ns]
Price             int64
dtype: object
```

Now, `Date` is a **datetime object** and supports time-based operations!

---

## ✅ **2.2 Extracting Date-Time Components**

Once converted, you can extract parts like **year, month, day**, etc.

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()

print(df)
```

✅ **Output:**
```
        Date  Price  Year  Month  Day  Weekday
0 2024-01-01    100  2024      1    1   Monday
1 2024-01-02    105  2024      1    2  Tuesday
2 2024-01-03    103  2024      1    3 Wednesday
```

---

## ✅ **2.3 Creating a Date Range**

You can **generate a full range of dates** using `pd.date_range()`:

```python
date_range = pd.date_range(start="2024-01-01", end="2024-01-07", freq='D')
print(date_range)
```

✅ **Output:**
```
DatetimeIndex(['2024-01-01', '2024-01-02', ..., '2024-01-07'], dtype='datetime64[ns]', freq='D')
```

📌 **Common Frequencies:**
- `"D"` = daily  
- `"M"` = month end  
- `"MS"` = month start  
- `"H"` = hourly  
- `"Q"` = quarterly  
- `"A"` = year end

---

## ✅ **2.4 Time Delta and Date Arithmetic**

You can calculate time differences using `Timedelta`:

```python
df["Days Since"] = df["Date"] - df["Date"].min()
print(df)
```

✅ **Output:**
```
        Date  Price Days Since
0 2024-01-01    100     0 days
1 2024-01-02    105     1 days
2 2024-01-03    103     2 days
```

---

## ✅ **2.5 Handling Custom Date Formats**

Sometimes, the date format may not be `YYYY-MM-DD`. You can specify a **custom format**:

```python
df_custom = pd.DataFrame({"date": ["01-03-2024", "02-03-2024"]})
df_custom["date"] = pd.to_datetime(df_custom["date"], format="%d-%m-%Y")
print(df_custom)
```

✅ **Output:**
```
        date
0 2024-03-01
1 2024-03-02
```

📌 **Common format codes:**

| Code | Meaning        |
|------|----------------|
| %d   | Day (01–31)    |
| %m   | Month (01–12)  |
| %Y   | Year (4-digit) |
| %b   | Month name (Jan–Dec) |

---

## ✅ **2.6 Setting Date as Index**

To make time-series analysis easier, we **set date as the index**:

```python
df.set_index("Date", inplace=True)
print(df)
```

✅ **Output:**
```
            Price  Year  Month  Day  Weekday  Days Since
Date                                                 
2024-01-01    100  2024      1    1   Monday     0 days
2024-01-02    105  2024      1    2  Tuesday     1 days
2024-01-03    103  2024      1    3 Wednesday     2 days
```

This allows for **easy slicing and resampling**, which we’ll explore next.

## 🟢 **3. Date-Time Indexing and Slicing in Pandas**

Understanding how to **index and slice** time-series data by date is crucial when analyzing trends and patterns in financial datasets.

---

## 🔹 **What Is Date-Time Indexing?**
When you set a **datetime column as the index**, Pandas lets you:
- **Slice** time-based rows (e.g., all data from 2022)
- Perform **resampling**, **rolling windows**, and **aggregation**
- Use **partial string indexing** like `'2022'` or `'2022-01'`

---

## ✅ **3.1 Set Date as Index (If Not Already Set)**

df = pd.DataFrame(data)

## 🟢 **4. Identifying Trends and Patterns in Time-Series Data**

---

### 🔹 **Why Identify Trends and Patterns?**

In financial time-series data, it’s essential to recognize:
- 📈 **Trends**: To understand long-term movement.
- 🔁 **Seasonality/Patterns**: To identify repeating behaviors (monthly, yearly).
- 📉 **Volatility and abrupt changes**: Helps detect anomalies and risks.

These insights help with:
- Better investment decisions.
- Early warnings for financial losses.
- Accurate forecasting and planning.

---

## ✅ **4.1 Rolling (Moving) Averages to Detect Trends**

**Rolling mean (or moving average)** smooths out short-term fluctuations and highlights long-term trends.

### 🔸 Example: 3-Day Moving Average
```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock price data
dates = pd.date_range("2024-01-01", periods=10, freq='D')
prices = [100, 102, 101, 105, 107, 106, 108, 110, 109, 112]
df = pd.DataFrame({"Price": prices}, index=dates)

# Calculate 3-day moving average
df["3-Day MA"] = df["Price"].rolling(window=3).mean()

# Plotting
df.plot(title="📈 Stock Price vs 3-Day Moving Average", figsize=(10,5))
plt.ylabel("Price (PKR)")
plt.grid(True)
plt.show()
```

✅ **What You’ll See:**
- A line for actual prices.
- A smoother line for the 3-day trend.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock price data
dates = pd.date_range("2024-01-01", periods=10, freq='D')
prices = [100, 102, 101, 105, 107, 106, 108, 110, 109, 112]
df = pd.DataFrame({"Price": prices}, index=dates)

# Calculate 3-day moving average
df["3-Day MA"] = df["Price"].rolling(window=3).mean()

# Plotting
df.plot(title="📈 Stock Price vs 3-Day Moving Average", figsize=(10,5))
plt.ylabel("Price (PKR)")
plt.grid(True)
plt.show()
````

## ✅ **4.3 Differencing to Remove Trend (`diff`)**

**Differencing** subtracts the previous value to remove trend and make the series **stationary** (useful in forecasting models).

```python
df["Difference"] = df["Price"].diff()
print(df[["Price", "Difference"]])
```

✅ Output:
```
            Price  Difference
2024-01-01    100         NaN
2024-01-02    102         2.0
2024-01-03    101        -1.0
...
```

## ✅ **4.4 Detecting Sudden Spikes or Drops**

You can detect **volatility or anomalies** using percent change or rolling std deviation:

```python
df["Rolling Std (3)"] = df["Price"].rolling(window=3).std()
print(df[["Price", "Rolling Std (3)"]])
```

✔ Helps to detect periods with **high variation or risk**.
