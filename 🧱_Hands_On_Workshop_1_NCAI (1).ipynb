## üì¶ **1. What is Warehouse Optimization?**

### üìñ **Definition**

Warehouse Optimization is the process of designing and managing warehouse operations in the **most efficient and cost-effective way**, while still ensuring speed, accuracy, and flexibility.
It covers everything from how goods are stored and retrieved, to how robots or workers move through the facility.

---

### üéØ **Goals of Warehouse Optimization**

1. **Better Space Utilization**

   * Maximize use of available floor and vertical space.
   * Reduce wasted aisles and unorganized shelves.

2. **Faster Picking Speed**

   * Improve how quickly items can be located, picked, and packed.
   * Use intelligent slotting (placing fast-moving items closer).

3. **Optimal Shelf Allocation**

   * Assign shelves based on product demand frequency.
   * Keep heavy or bulky items in easy-to-reach spots.

4. **Lower Operational Costs**

   * Reduce unnecessary worker travel time.
   * Cut down on handling and storage costs.

5. **Error Reduction & Safety**

   * Minimize picking errors.
   * Reduce workplace hazards by better planning of traffic flow.

---

## ‚ö†Ô∏è **2. Common Bottlenecks in Traditional Warehouses**

Despite being the backbone of supply chains, traditional warehouses face several **bottlenecks (slowdowns/inefficiencies)**:

1. **Poor Layout & Space Wastage**

   * Shelves not arranged based on demand patterns.
   * Empty spaces in racks or cluttered aisles.

2. **Slow Picking Process**

   * Workers spend too much time walking between shelves.
   * Lack of intelligent product placement.

3. **Inventory Inaccuracy**

   * Manual entry errors leading to stockouts or overstocking.
   * Mismatch between physical stock and system records.

4. **Limited Visibility**

   * No real-time tracking of goods.
   * Managers lack live insights on stock levels and bottlenecks.

5. **High Labor Dependency**

   * Processes rely heavily on manual work ‚Üí prone to fatigue, mistakes, and high costs.

6. **Poor Order Fulfillment Speed**

   * Difficulty in keeping up with **e-commerce demand** (like same-day or next-day delivery).

---

‚úÖ Example: A local **grocery chain warehouse** may store sugar, rice, and oil randomly across faraway shelves. Workers walk extra distances to collect them for each order, slowing down fulfillment and raising costs. Optimized placement could drastically cut travel time.

# ü§ñ **AI in Warehouse Management**

AI transforms warehouses from being **reactive storage centers** into **smart, predictive, and automated ecosystems**. Instead of waiting for demand to happen, AI helps warehouses *anticipate, plan, and act* in real-time.

---

## üõ£Ô∏è **1. Role of AI in Route Planning, Shelf Allocation & Demand-Based Stocking**

* **Route Planning**
  AI systems calculate the **shortest, least-congested routes** for workers or robots.

  * Example: A picker needs 10 items ‚Üí AI arranges the route to minimize walking distance.
  * Algorithms like **Dijkstra‚Äôs shortest path** or reinforcement learning can power this.

* **Shelf Allocation (Slotting Optimization)**
  AI decides where each product should be placed for fastest access.

  * **Fast-moving items** (e.g., milk, bread in a grocery warehouse) are placed near dispatch zones.
  * **Seasonal items** (e.g., fans in summer, heaters in winter) are rotated closer when needed.

* **Demand-Based Stocking**
  AI predicts demand surges using historical data + trends.

  * Example: Eid, Black Friday, or Ramadan ‚Üí AI increases stock of high-demand items.
  * Prevents both *overstocking* and *stockouts*.

---

## üß© **2. Clustering Algorithms for Zone Picking**

* **Problem**: Workers waste time zig-zagging across the warehouse.
* **AI Solution**: Use **clustering algorithms (like K-Means)** to group items into *zones*.

  * Zone Picking = Each worker/robot is responsible for a **specific area/cluster**.
  * Example:

    * Worker A handles **Zone 1 (drinks, snacks)**
    * Worker B handles **Zone 2 (stationery, electronics)**

This reduces congestion and increases picking speed.

---

## ‚ö° **3. Real-Time Inventory Decision-Making**

* AI integrates with **IoT sensors, RFID tags, and ERP systems** to provide *live updates*.
* Instead of monthly audits, managers see:

  * üìâ *Low-stock alerts* (trigger automatic reorders).
  * üì¶ *Overstock alerts* (suggest discounts/promotions).
  * üïí *Dynamic reallocation* (if product A isn‚Äôt selling, move it further back, bring fast-selling B forward).

---

## üëÄ **4. Computer Vision for Inventory Tracking**

AI doesn‚Äôt just rely on humans scanning barcodes ‚Äî it **sees** the warehouse.

* **Barcode and QR Scanning Automation**
  Handheld or mounted scanners automatically update stock when items move.

* **Object Detection with Cameras**
  Cameras + AI (like **YOLOv8, OpenCV**) detect products, boxes, or even misplaced pallets in real-time.

  * Example: A camera above a shelf can alert if a box is missing or wrongly placed.
  * Reduces dependency on manual checks.

---

## üöÅ **5. Drones & Robots in Warehouses**

* **Drones** fly across racks to scan inventory faster than humans.
* **Autonomous Mobile Robots (AMRs)** carry goods to packing stations.

  * Example: **Amazon‚Äôs Kiva robots** reduce walking distance for workers by 70‚Äì80%.

These AI-powered machines boost speed, accuracy, and safety while lowering labor costs.

---

‚úÖ *Bottom line*: AI makes warehouses **predictive, efficient, and automated** ‚Äî a shift from *manual chaos* to *smart orchestration*.

# üåê **AI + IoT in Logistics**

---

* **RFID Tags & IoT Sensors**

  * Location ‚Üí Where exactly is the truck/pallet now?
  * Condition ‚Üí Is the product being stored correctly (fragile, upright, sealed)?

* **AI Integration**

  * Alert managers when goods are delayed or diverted.
  * Automatically update customers about estimated delivery time.

‚úÖ Example: *Courier services like TCS/DHL use RFID + GPS IoT sensors for live parcel tracking.*

## üõ£Ô∏è **2. Predictive Route Planning (Traffic + Weather Aware)**

* **Problem**: Traditional route planning only considers distance.

* **AI + IoT Solution**:

  * IoT devices on trucks ‚Üí collect live road data (speed, traffic jams, accidents).
  * AI models ‚Üí predict **best alternate routes**.
  * Weather APIs ‚Üí suggest avoiding flood-prone areas or icy roads.

* **Benefit**: Reduces delays, fuel cost, and ensures on-time delivery.

‚úÖ Example: *Foodpanda or Careem Delivery uses AI-based algorithms to reroute riders dynamically during traffic congestion.*

---

## ‚ùÑÔ∏è **3. Cold Chain Logistics with Temperature Sensors**

* **IoT Sensors** constantly monitor:

  * Temperature üå°Ô∏è
  * Humidity üíß
  * Door opening/closing events üö™

* **AI Systems** analyze this sensor data to:

  * Trigger alerts if temperature goes above threshold (e.g., milk shipment warming up).
  * Suggest preventive actions (reroute to nearest cold storage).
  * Predict risks in advance (e.g., power failure at cold depot).

‚úÖ Example: *Pfizer vaccine distribution relied on IoT temperature sensors + AI dashboards to maintain ultra-cold storage during COVID-19 supply chains.*

## üß≠ PART 1: Route Optimization (Delivery Route Problem)

### üß© Problem:

You have one warehouse and multiple delivery locations. You want to **visit every customer** and come back - but with **minimum travel distance** (save fuel, save time!).

### üß† How AI helps:

We use a simple AI technique called **Greedy Algorithm**. It works like this:

* Start at the warehouse.
* Find the **closest customer**.
* Go there.
* From there, find the next closest one...
* Keep repeating until all are visited.
* Come back to warehouse.

üìå This mimics how a human would do it if they just kept choosing the "next nearest" place.

### üìä What we see:

* A **list** showing the optimized delivery route (e.g., Warehouse ‚Üí A ‚Üí B ‚Üí C ‚Üí D ‚Üí back to Warehouse).

### üß† How AI helps:

We use a popular AI method called **K-Means Clustering**. It:

* Looks at each product‚Äôs **demand (how often people buy it)** and **size**.
* Groups them into **zones** (like Shelf 1, Shelf 2, Shelf 3).
* Puts **similar items together** so the warehouse is neat and fast to access.

````python
# ================================================================
# üöö WEEK 11 ‚Äì DAY 2 LAB
# AI-Powered Warehouse Optimization & Logistics Automation
# ================================================================
# This notebook covers 3 tasks:
# 1. Route Optimization (Travelling Salesman Problem - TSP)
# 2. Shelf Space Allocation using K-Means Clustering
# 3. Simulated Inventory Tracker with Pandas
# ================================================================
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# ======================
# üì¶ Import Libraries
# ======================

# Numpy: for numerical operations (distance calculations, arrays)
import numpy as np

# Pandas: for handling tabular data like products/inventory
import pandas as pd

# Matplotlib + Seaborn: for creating plots and visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# KMeans: clustering algorithm from scikit-learn
from sklearn.cluster import KMeans

# ipywidgets: for interactive search dropdowns (only works in Jupyter/Colab)
try:
    from ipywidgets import interact
except:
    pass


# ================================================================
# üß≠ PART 1: Simulated Route Optimization (TSP)
# ================================================================
print("\n" + "="*70)
print("üß≠ PART 1: SIMULATED ROUTE OPTIMIZATION (Travelling Salesman Problem)")
print("="*70)

# Create dictionary of delivery locations with (x, y) coordinates
# Warehouse is starting point, Customers A‚ÄìD are destinations
locations = {
    "Warehouse": (0, 0),
    "Customer A": (2, 4),
    "Customer B": (5, 2),
    "Customer C": (6, 6),
    "Customer D": (8, 3)
}

# Convert dictionary values to numpy array for easier math operations
points = np.array(list(locations.values()))

# Extract the location names (Warehouse, Customer A, etc.)
labels = list(locations.keys())

# Function: Greedy Nearest Neighbor algorithm for TSP
def tsp_greedy(points, start=0):
    n = len(points)                      # number of locations
    visited = [False] * n                # track visited locations
    route = [start]                      # start route from 'start' index
    visited[start] = True                # mark starting point visited
    total_dist = 0                       # total distance traveled

    # Loop until all locations are visited
    for _ in range(n - 1):
        last = route[-1]                 # current location index
        nearest = None                   # nearest neighbor index
        nearest_dist = float('inf')      # initialize with infinity

        # Check all other locations
        for i in range(n):
            if not visited[i]:           # if location not visited
                # Euclidean distance between current and candidate point
                dist = np.linalg.norm(points[last] - points[i])
                if dist < nearest_dist:  # update nearest if closer
                    nearest = i
                    nearest_dist = dist

        # Add nearest neighbor to route
        route.append(nearest)
        visited[nearest] = True
        total_dist += nearest_dist       # add to total distance

    # Finally, return to starting point
    total_dist += np.linalg.norm(points[route[-1]] - points[start])
    route.append(start)                  # close the loop

    return route, total_dist

# Run algorithm on our dataset
route, total_distance = tsp_greedy(points, start=0)

# Print the optimized route
print("\nOptimized Delivery Route:")
for idx in route:
    print(" ‚Üí", labels[idx], end="")
print(f"\n\nüìè Total Travel Distance: {round(total_distance, 2)} units")

# Visualization of optimized path
plt.figure(figsize=(6,6))
# Draw dashed lines between connected points in route
for i in range(len(route)-1):
    p1, p2 = points[route[i]], points[route[i+1]]
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b--')

# Plot locations as red dots
plt.scatter(points[:,0], points[:,1], c='red', s=100)

# Annotate each point with its label
for i, label in enumerate(labels):
    plt.text(points[i][0]+0.1, points[i][1]+0.1, label, fontsize=12)

plt.title("üß≠ Optimized Delivery Route (Greedy TSP)")
plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")
plt.show()


# ================================================================
# üìä PART 2: Shelf Space Allocation using K-Means Clustering
# ================================================================
print("\n" + "="*70)
print("üìä PART 2: SHELF SPACE ALLOCATION (K-Means Clustering)")
print("="*70)

# Create a simulated dataset of products
# Features: Demand_Frequency, Size, Category_Code
data = {
    'Product': ['A','B','C','D','E','F','G','H','I','J'],
    'Demand_Frequency': [200, 50, 300, 120, 80, 400, 20, 180, 260, 95],
    'Size': [3, 2, 5, 4, 3, 6, 2, 4, 5, 3],
    'Category_Code': [1, 2, 1, 3, 2, 1, 3, 2, 1, 3]
}
df = pd.DataFrame(data)

# Print original product table
print("\nüîç Original Product Data")
print(df.to_string(index=False))

# Select features for clustering
X = df[['Demand_Frequency', 'Size']]

# Apply K-Means clustering with 3 shelf zones
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Shelf_Zone'] = kmeans.fit_predict(X)

# Print products with assigned shelf zones
print("\nüì¶ Products Assigned to Shelf Zones")
print(df[['Product','Demand_Frequency','Size','Shelf_Zone']].to_string(index=False))

# Visualization of clustering results
plt.figure(figsize=(8,6))
# Scatterplot: Demand vs Size, colored by Shelf_Zone
sns.scatterplot(data=df, x='Demand_Frequency', y='Size',
                hue='Shelf_Zone', palette='Set1', s=120)

# Annotate product IDs on the scatterplot
for i, row in df.iterrows():
    plt.text(row['Demand_Frequency']+5, row['Size']+0.1, row['Product'], fontsize=9)

plt.title("üìä Shelf Allocation Clusters (K-Means)")
plt.xlabel("Demand Frequency")
plt.ylabel("Product Size")
plt.legend(title="Shelf Zone")
plt.show()


# ================================================================
# üì¶ PART 3: Simulated Inventory Tracker
# ================================================================
print("\n" + "="*70)
print("üì¶ PART 3: SIMULATED INVENTORY TRACKER")
print("="*70)

# Create inventory dataset
inventory_data = {
    "Product_ID": ["P101", "P102", "P103", "P104", "P105"],
    "Product_Name": ["Sugar", "Rice", "Oil", "Tea", "Biscuits"],
    "Location": ["Shelf A", "Shelf B", "Shelf C", "Shelf D", "Shelf A"],
    "Status": ["IN", "IN", "OUT", "IN", "IN"],
    "Quantity": [50, 30, 0, 20, 100]
}
df_inv = pd.DataFrame(inventory_data)

# Print initial inventory
print("\nüìã Current Inventory")
print(df_inv.to_string(index=False))

# Function to update stock levels
def update_stock(product_id, change, status):
    # Check if product exists in inventory
    if product_id in df_inv["Product_ID"].values:
        # Get row index
        idx = df_inv[df_inv["Product_ID"] == product_id].index[0]
        # Update quantity (+ for incoming, - for outgoing)
        df_inv.at[idx, "Quantity"] += change
        # Update status (IN/OUT)
        df_inv.at[idx, "Status"] = status
    else:
        print("‚ùå Product not found!")

# Simulate transactions
update_stock("P101", -10, "OUT")   # Sugar: 10 units OUT
update_stock("P102", +15, "IN")    # Rice: 15 units IN

# Print updated inventory after changes
print("\nüì¶ Updated Inventory After Transactions")
print(df_inv.to_string(index=False))

# Optional interactive search dropdown (works in Jupyter/Colab)
try:
    @interact(Product_ID=df_inv["Product_ID"].tolist())
    def search_inventory(Product_ID):
        display(df_inv[df_inv["Product_ID"] == Product_ID])
except:
    pass
````

## üöö **4Ô∏è‚É£ Route Optimization (Travelling Salesman Problem)**

This module simulates how a delivery van visits multiple customer locations with **minimum travel distance**. We'll use a simple greedy algorithm to solve the **TSP** (Travelling Salesman Problem).

---

### üßæ üìÑ Step-by-Step User Flow

1. App displays sample map (Warehouse + Customers A‚ÄìD).
2. User can **see the optimized delivery route**.
3. Visualizes the route path (as a line plot).
4. **Later** (optional): User can **upload custom locations**.

### ‚úÖ Full Working Code (Add under `if tab == "Route Optimization üß≠":`)

```python
import numpy as np                       # For numerical operations (like distance)
import matplotlib.pyplot as plt          # For route visualization

if tab == "Route Optimization üß≠":
    st.subheader("üß≠ Delivery Route Optimizer")  # Section heading

    # ---------------------------------------------
    # Step 1: Define default delivery coordinates
    # ---------------------------------------------
    st.markdown("#### üìç Default Delivery Points")
    st.info("Warehouse and 4 Customer locations")

    # Dictionary with names and (x, y) coordinates
    locations = {
        "Warehouse": (0, 0),
        "Customer A": (2, 4),
        "Customer B": (5, 2),
        "Customer C": (6, 6),
        "Customer D": (8, 3)
    }

    # Convert location values to NumPy array for easier math
    points = np.array(list(locations.values()))

    # Extract the names (for labeling route)
    labels = list(locations.keys())

    # ---------------------------------------------
    # Step 2: Define the TSP Greedy Algorithm
    # ---------------------------------------------
    def tsp_greedy(points, start=0):
        n = len(points)                      # Total number of locations
        visited = [False] * n                # Track which locations are visited
        route = [start]                      # Start from 'Warehouse'
        visited[start] = True                # Mark it as visited
        total_dist = 0                       # To store total distance

        # Loop through all other locations
        for _ in range(n - 1):
            last = route[-1]                 # Last visited location index
            nearest = None                   # Will store the nearest neighbor index
            nearest_dist = float('inf')      # Set initial distance to infinity

            # Check all other unvisited locations
            for i in range(n):
                if not visited[i]:           # Skip if already visited
                    # Euclidean distance from last location to current one
                    dist = np.linalg.norm(points[last] - points[i])
                    if dist < nearest_dist:  # If it's the closest so far
                        nearest = i
                        nearest_dist = dist

            # Add the nearest neighbor to the route
            route.append(nearest)
            visited[nearest] = True
            total_dist += nearest_dist       # Add distance to total

        # Return to starting point (Warehouse)
        total_dist += np.linalg.norm(points[route[-1]] - points[start])
        route.append(start)                  # Close the route

        return route, total_dist             # Return final route and distance

    # ---------------------------------------------
    # Step 3: Call the algorithm and show route
    # ---------------------------------------------
    route, total_distance = tsp_greedy(points)

    # Show route as text
    st.markdown("#### üó∫Ô∏è Optimized Route")
    route_str = " ‚Üí ".join([labels[i] for i in route])  # Convert indices to names
    st.success(f"**Optimized Route:** {route_str}")     # Display route
    st.info(f"üìè Total Travel Distance: {round(total_distance, 2)} units")  # Show distance

    # ---------------------------------------------
    # Step 4: Visualize the route with Matplotlib
    # ---------------------------------------------
    fig, ax = plt.subplots()                  # Create a figure and axis

    # Draw lines between each connected point in the route
    for i in range(len(route) - 1):
        p1, p2 = points[route[i]], points[route[i + 1]]   # Consecutive points
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b--')     # Dashed blue line

    # Plot the delivery locations
    ax.scatter(points[:, 0], points[:, 1], c='red', s=100)  # Red dots for locations

    # Add labels (Warehouse, Customer A, etc.)
    for i, label in enumerate(labels):
        ax.text(points[i][0] + 0.1, points[i][1] + 0.1, label, fontsize=10)

    # Add chart title and axes labels
    ax.set_title("üß≠ Optimized Delivery Route (TSP)")
    ax.set_xlabel("X Coordinate")
    ax.set_ylabel("Y Coordinate")

    # Show plot in Streamlit
    st.pyplot(fig)

```

````python
# ================================================================
# üöö AI-Powered Warehouse Optimization App (Streamlit)
# ================================================================
# üìç Route Optimization (Travelling Salesman Problem)
# üìä Shelf Space Allocation (K-Means Clustering)
# üì¶ Inventory Management with Transactions
# üìÅ CSV Upload Support
# ================================================================

import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.cluster import KMeans

# Set Streamlit page configuration
st.set_page_config(page_title="Warehouse AI Optimizer", layout="centered")

# Sidebar for navigation
st.sidebar.title("üß≠ Warehouse AI App")
tab = st.sidebar.radio("Go to", ["üè† Home", "üß≠ Route Optimizer", "üìä Shelf Clustering", "üì¶ Inventory Tracker"])

# ---------------------------------------------
# üè† HOME
# ---------------------------------------------
if tab == "üè† Home":
    st.title("üöö AI-Powered Warehouse Optimization")
    st.markdown("""
    This interactive app demonstrates how AI can optimize:
    - **Delivery Routes** (TSP)
    - **Shelf Space Allocation** (K-Means Clustering)
    - **Inventory Tracking** (Real-time updates)

    Built with **Python**, **Pandas**, **Scikit-learn**, **Matplotlib**, and **Streamlit**.
    """)

# ---------------------------------------------
# üß≠ ROUTE OPTIMIZER (Travelling Salesman Problem)
# ---------------------------------------------
elif tab == "üß≠ Route Optimizer":
    st.title("üß≠ Route Optimizer ‚Äì TSP")

    # Coordinates for warehouse and customers
    locations = {
        "Warehouse": (0, 0),
        "Customer A": (2, 4),
        "Customer B": (5, 2),
        "Customer C": (6, 6),
        "Customer D": (8, 3)
    }

    # Extract coordinates and labels
    points = np.array(list(locations.values()))
    labels = list(locations.keys())

    # TSP algorithm (Greedy Nearest Neighbor)
    def tsp_greedy(points, start=0):
        n = len(points)
        visited = [False] * n
        route = [start]
        visited[start] = True
        total_dist = 0
        for _ in range(n - 1):
            last = route[-1]
            nearest = None
            nearest_dist = float('inf')
            for i in range(n):
                if not visited[i]:
                    dist = np.linalg.norm(points[last] - points[i])
                    if dist < nearest_dist:
                        nearest = i
                        nearest_dist = dist
            route.append(nearest)
            visited[nearest] = True
            total_dist += nearest_dist
        total_dist += np.linalg.norm(points[route[-1]] - points[start])
        route.append(start)
        return route, total_dist

    route, total_distance = tsp_greedy(points)

    # Show Route
    st.subheader("üöö Optimized Route")
    for idx in route:
        st.write("‚û°Ô∏è", labels[idx])

    st.write(f"üìè **Total Distance:** `{round(total_distance, 2)} units`")

    # Visualize Route
    st.subheader("üìç Route Map")
    fig, ax = plt.subplots(figsize=(6, 6))
    for i in range(len(route) - 1):
        p1, p2 = points[route[i]], points[route[i+1]]
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b--')
    ax.scatter(points[:, 0], points[:, 1], c='red', s=100)
    for i, label in enumerate(labels):
        ax.text(points[i][0]+0.1, points[i][1]+0.1, label, fontsize=12)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("Optimized Delivery Route")
    st.pyplot(fig)

# ---------------------------------------------
# üìä SHELF CLUSTERING (K-Means)
# ---------------------------------------------
elif tab == "üìä Shelf Clustering":
    st.title("üìä Shelf Space Allocation ‚Äì K-Means")

    # Sample product data
    data = {
        'Product': ['A','B','C','D','E','F','G','H','I','J'],
        'Demand_Frequency': [200, 50, 300, 120, 80, 400, 20, 180, 260, 95],
        'Size': [3, 2, 5, 4, 3, 6, 2, 4, 5, 3],
    }
    df = pd.DataFrame(data)

    st.subheader("üìã Product Table")
    st.dataframe(df)

    # Clustering
    X = df[['Demand_Frequency', 'Size']]
    kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
    df['Shelf_Zone'] = kmeans.fit_predict(X)

    st.subheader("üì¶ Shelf Zones")
    st.dataframe(df)

    # Plot
    fig2 = px.scatter(df, x='Demand_Frequency', y='Size',
                      color='Shelf_Zone', text='Product',
                      title="Shelf Allocation (K-Means)",
                      color_continuous_scale='Portland')
    st.plotly_chart(fig2)

# ---------------------------------------------
# üì¶ INVENTORY TRACKER (with CSV Upload)
# ---------------------------------------------
elif tab == "üì¶ Inventory Tracker":
    st.title("üì¶ Inventory Tracker")

    st.markdown("#### üìÅ Upload Inventory CSV (Optional)")
    uploaded_file = st.file_uploader("Choose CSV file", type="csv")

    if uploaded_file is not None:
        df_inv = pd.read_csv(uploaded_file)
    else:
        df_inv = pd.DataFrame({
            "Product_ID": ["P101", "P102", "P103", "P104", "P105"],
            "Product_Name": ["Sugar", "Rice", "Oil", "Tea", "Biscuits"],
            "Location": ["Shelf A", "Shelf B", "Shelf C", "Shelf D", "Shelf A"],
            "Status": ["IN", "IN", "OUT", "IN", "IN"],
            "Quantity": [50, 30, 0, 20, 100]
        })

    st.markdown("#### üìã Current Inventory")
    st.dataframe(df_inv)

    st.markdown("#### üîÅ Update Stock")

    with st.form("stock_form"):
        product_id = st.selectbox("Select Product", df_inv["Product_ID"])
        change = st.number_input("Quantity Change (+/-)", value=0, step=1)
        status = st.selectbox("Status", ["IN", "OUT"])
        update = st.form_submit_button("Apply Update")

        if update:
            idx = df_inv[df_inv["Product_ID"] == product_id].index[0]
            df_inv.at[idx, "Quantity"] += change
            df_inv.at[idx, "Status"] = status
            st.success(f"‚úÖ Updated stock for {product_id}")

    st.markdown("#### üì¶ Updated Inventory")
    st.dataframe(df_inv)

    st.markdown("#### üìä Inventory Bar Chart")
    fig3 = px.bar(df_inv, x="Product_Name", y="Quantity", color="Status",
                  title="Stock Level per Product", text="Quantity")
    st.plotly_chart(fig3)

````
