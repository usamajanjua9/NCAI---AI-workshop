# 📌 **1. What is Demand Forecasting?**

---

### 📖 **Definition**

Demand forecasting is the process of **predicting how much of a product or service people will want in the future**. This prediction is based on different types of data like:

* Past sales records
* Seasonal trends (like more ice cream in summer)
* Holidays or festivals
* Price changes
* Market trends
* Economic conditions (e.g., inflation, fuel prices)

This helps businesses **prepare in advance** so they don’t run out of products or have too many left over.
<img src="https://www.scdigest.com/images/Cartoons/16-05-31winner1a.gif" width="350" >

---

### 🎯 **Why Is Demand Forecasting Important in Supply Chains?**

A supply chain includes everything from producing a product to delivering it to customers. Forecasting helps each step work smoothly. Here’s how:

| Area                      | Benefit of Forecasting                                                       |
| ------------------------- | ---------------------------------------------------------------------------- |
| 🏭 **Production**         | Makes sure the right quantity is manufactured at the right time              |
| 📦 **Inventory**          | Helps avoid too much stock (extra storage cost) or too little (missed sales) |
| 🚚 **Delivery/Logistics** | Ensures enough trucks and delivery staff are ready when demand rises         |
| 👩‍💼 **Finance**         | Supports better budgeting and cash flow decisions                            |
| 👥 **Customer Service**   | Reduces delays and increases satisfaction by having products ready           |

> ⚠️ **If a business does not forecast demand properly**, it may:
>
> * Waste money on unsold products
> * Lose customers due to unavailability
> * Have chaos in warehouses and production lines

## 📌 **2. Types of Demand Forecasting**

---

Forecasting is not one-size-fits-all. There are **different ways to forecast**, based on:

* **Available data**
* **Type of product**
* **Timeframe**
* **Market behavior**

Let’s explore the main types with **simple explanations and examples**.

### 🧪 **2.1 Qualitative Forecasting**

🧠 **“Guess with experience”**

This method uses **human judgment** and **opinions of experts** to estimate future demand. It doesn’t rely on numbers but on:

* Surveys
* Focus groups
* Expert panels
* Customer feedback

#### ✅ **When is it used?**

* For **new products** with no sales history
* In **uncertain times** (pandemics, elections, natural disasters)
* When entering **new markets**

#### ✅ **Real-World Examples**

| Company               | Use Case                                                                                                      |
| --------------------- | ------------------------------------------------------------------------------------------------------------- |
| 👟 **Nike**           | Predicts demand for a new shoe model by asking store managers and running online polls                        |
| 🍔 **New Restaurant** | Owners estimate how many meals to prepare on opening day based on neighborhood habits and competitor analysis |
| 📱 **Apple**          | Combines expert advice and design trends to forecast demand for a brand-new iPhone feature                    |

📌 *Note: This type can be risky since it’s not based on data – but it’s still very helpful in early stages.*

### 📊 **2.2 Quantitative Forecasting**

📉 **“Guess using numbers and data”**

#### 🔢 Common tools:

* Moving Averages
* Regression models
* Time series analysis
* Machine Learning (Random Forest, XGBoost, etc.)

#### ✅ **When is it used?**

* When you have **a lot of past sales data**
* For **regular, repeating products** like groceries or medicines
* To create **automated dashboards and models**

# 📆 **5. Seasonality & Trend Detection**

Understanding **patterns in time-based data** is crucial before forecasting. AI models do this using trend and seasonality analysis.

---


<img src="https://www.colibritrader.com/wp-content/uploads/2021/11/ezgif.com-gif-maker-9.gif" width="350" >

### 🔄 **5.1 Moving Averages**

📌 A **moving average** smooths out short-term fluctuations and shows long-term trends.

* **Simple Moving Average (SMA):** Average over a fixed time window
* **Use Case:**

  * To identify general **upward or downward trend**
  * Remove noise caused by promotions or one-day events

#### ✅ Example:

> A 7-day moving average helps a grocery store see weekly sales patterns while ignoring individual daily spikes.

---
<img src="https://ulvgard.se/img/articles/trend_and_seasonality_decomposition_with_prophet/prophet_triangle_noise.gif" width="450" >

### 🧮 **5.2 Time Series Decomposition**

Time series data can be **broken into three parts**:

1. **Trend** – General direction (up/down)
2. **Seasonality** – Repeating pattern (e.g., higher sales on weekends)
3. **Residual** – Irregular/unpredictable noise

> This helps us understand **what part of the demand is seasonal** and **what is truly changing**.

#### ✅ Example:

> Ice cream sales spike every summer (seasonality), but this year they are even higher than usual (positive trend).

---

<img src="https://portent.com/conversationmarketing/rolling-average/rolling-average-graph.gif" width="400" >

### 📉 **5.3 Rolling Mean & Standard Deviation**

* **Rolling Mean**: Like a moving average but recalculated at every time step
* **Rolling Std Dev**: Shows how much demand is varying over time

> 🔍 Helps detect:
>
> * Periods of **stable** vs **highly volatile** demand
> * Sudden **shocks** like market disruptions or panic buying

#### ✅ Example:

> Rolling standard deviation shows a spike in demand variation for hand sanitizers during the early COVID-19 period.

---

📊 **Mini Example:**

| Date  | Sales | 3-Day Rolling Mean      |
| ----- | ----- | ----------------------- |
| Jan 1 | 100   | -                       |
| Jan 2 | 120   | -                       |
| Jan 3 | 130   | (100+120+130)/3 = 116.7 |
| Jan 4 | 140   | (120+130+140)/3 = 130.0 |

This helps businesses understand where **seasonal promotions** or **abnormal demand** may affect future planning.

## 🔢 **6. Common AI/ML Algorithms Used in Demand Forecasting**

Machine Learning (ML) offers a variety of algorithms for predicting demand. Each has strengths depending on the **type of data**, **pattern complexity**, and **business needs**.

### 📈 **6.1 Linear Regression**

---

### 🧠 What Is It?

Linear Regression is the **simplest forecasting model**.
It assumes that the relationship between the input (like day, month, price) and the output (sales/demand) is **linear** — meaning it can be drawn as a straight line.

### 🧪 Formula:

$$
\text{Predicted Demand (Y)} = m \cdot x + b
$$

Where:

* `x` is your input feature (e.g., day number)
* `m` is the slope (rate of change)
* `b` is the intercept (starting value)

<img src="https://miro.medium.com/v2/resize:fit:612/1*eRHWTUi-OLoFF6bZEfcAAw.png" width="400" >

### ✅ Best For:

* Identifying **trend over time**
* Predicting **steady** increases or decreases in demand

### 🚫 Limitations:

* Can’t handle **seasonal patterns**, spikes, or complex relationships
* Works poorly with **nonlinear** data

### 🛍️ Real-World Example:

> A bookstore uses linear regression to predict daily sales based on the number of students on campus during the semester.

---

<img src="https://datasciencedojo.com/wp-content/uploads/random-forest-algorithm-random-forest.webp" width="400" >

### 🌳 **6.2 Decision Trees & Random Forest**

---

#### 🌴 **Decision Trees**

* These models make decisions like **yes/no questions**.
* For example:
  "Is it the weekend?" → "Yes" → "Expect more sales"

#### 🌳 **Random Forest**

* It’s an **ensemble** method that builds **many trees** and takes the average of their results.
* More accurate than a single tree and avoids overfitting.

### ✅ Best For:

* Handling **categorical + numerical features**
* Capturing **nonlinear relationships**
* Handling **missing data** and **noisy datasets**

### 🧠 How It Works:

* Splits data into different “branches” based on the best conditions (e.g., day of week, promotions)
* Learns patterns from past outcomes
* Makes predictions by combining decisions from multiple trees

### 🛒 Real-World Example:

> A supermarket uses a random forest to forecast how promotions, holidays, and temperature affect ice cream sales across cities.

---

---

Time series forecasting requires models that **understand order in data** (i.e., sales today depend on what happened yesterday).

#### 🧮 ARIMA (Auto-Regressive Integrated Moving Average)

* A **statistical** model, good for **univariate** time series (one variable like sales over time).
* It uses:

  * **Auto-regression** (past values)
  * **Differencing** (to remove trend)
  * **Moving averages** (smooth errors)

### ✅ Best For:

* Clean, **non-seasonal** data with clear trends
* Smaller datasets

### 🚫 Limitations:

* Needs **manual tuning**
* Can’t handle **complex external factors** like promotions or weather

#### ✅ Example:

> A utility company forecasts electricity usage using ARIMA based only on past consumption data.

#### 🧠 LSTM (Long Short-Term Memory)

* A **deep learning** model that remembers sequences over time.
* Part of **Recurrent Neural Networks (RNNs)**.
* Can model **long-term dependencies** in time series data.

### ✅ Best For:

* **Highly complex patterns** with seasonality, sudden changes, multiple inputs
* Large datasets

### 🚫 Limitations:

* Requires **a lot of data** and computing power
* Hard to interpret
* Training takes time

#### ✅ Example:

> Amazon uses LSTM to forecast demand for millions of products using search trends, weather, stock levels, and sales history.

### 🧙 **6.4 Prophet (by Facebook/Meta)**

---

### 🧠 What Is Prophet?

Prophet is an **open-source forecasting library** developed by Facebook.
It is designed to handle **daily time series data** that includes:

* **Trends**
* **Seasonality**
* **Holidays**
* **Missing data**

It works well **out-of-the-box** and is easy to tune.

### ✅ Best For:

* Business analysts without deep ML knowledge
* Datasets with **strong seasonal effects** (weekly, yearly, etc.)
* Long-term forecasting with **holiday effects**

### 🚫 Limitations:

* Doesn’t perform well on **very short or highly noisy datasets**
* Limited flexibility compared to full ML pipelines

#### ✅ Real-World Example:

> A retail chain uses Prophet to forecast daily sales while considering Eid, Independence Day, and other public holidays as spikes in demand.

## ⚖️ **7.1 Overstocking vs Understocking**

| Scenario            | Description                                            | Problem                                                                          |
| ------------------- | ------------------------------------------------------ | -------------------------------------------------------------------------------- |
| 📦 **Overstocking** | Having **too much inventory** in storage               | - High storage costs<br>- Product spoilage (for perishables)<br>- Wasted capital |
| ❌ **Understocking** | Having **too little inventory** when customers need it | - Lost sales<br>- Poor customer experience<br>- Emergency reorders (higher cost) |

## 📐 **7.2 EOQ (Economic Order Quantity)**

### 📖 Definition:

EOQ is the **ideal quantity** of a product that a company should order each time to minimize the **total cost of inventory**, which includes:

* Ordering cost (placing a purchase order)
* Holding cost (storing the product)
* Shortage cost (lost sales or rush orders)

---

### 🔢 EOQ Formula:

$$
EOQ = \sqrt{\frac{2DS}{H}}
$$

Where:

* $D$ = Annual demand (units)
* $S$ = Ordering cost per order
* $H$ = Holding cost per unit per year

---

### ✅ Example:

> Suppose a store sells 10,000 notebooks/year.
>
> * Ordering cost = Rs. 500/order
> * Holding cost = Rs. 2/notebook/year
>
> $$
> $$

$$
EOQ = \sqrt{\frac{2 \times 10,000 \times 500}{2}}
= \sqrt{5,000,000}
\approx 2236 \;\text{notebooks/order}
$$


This means ordering around 2,236 notebooks each time minimizes total cost.

---

### 🤖 How AI Helps with EOQ:

* Adjusts $D$, $S$, and $H$ **automatically** using real-time data
* Includes **seasonal demand**, **price fluctuations**, and **lead time variability**
* Suggests **dynamic EOQ** instead of fixed formulas

---

## ⏰ **7.3 Reorder Point Prediction**

The **reorder point (ROP)** is the inventory level at which you must place a **new order** so the stock doesn’t run out before the next delivery arrives.

---

### 📌 ROP Formula (Simple):

$$
ROP = \text{Average Daily Demand} \times \text{Lead Time}
$$


You can also add a **Safety Stock**:

$$
ROP = (\text{Daily Demand} \times \text{Lead Time}) + \text{Safety Stock}
$$

---

### ✅ Example:

> * Daily demand: 50 units
> * Lead time: 7 days
> * Safety stock: 100 units
>


$$
ROP = (50 \times 7) + 100 = 450 \text{ units}
$$

So, when inventory reaches 450, the system should place a new order.

---

### 🔍 AI-Powered ROP Prediction:

* Predicts future demand **based on ML models**, not just averages
* Dynamically adjusts lead times based on **supplier performance**, **weather**, or **holidays**
* Suggests **automated alerts** or even **automatic purchase orders**

---

### 🛒 Real-World Use Cases:

| Company                  | How They Use AI                                                                                                  |
| ------------------------ | ---------------------------------------------------------------------------------------------------------------- |
| 🏪 **7-Eleven**          | Predicts reorder points for each item daily using local weather, foot traffic, and sales patterns                |
| 💻 **HP**                | Optimizes printer cartridge inventory across warehouses using AI-based EOQ and reorder point strategies          |
| 🛍️ **Daraz.pk (Local)** | Can use AI to predict stock needs during flash sales to automatically trigger restocking before products run out |

| Concept | Formula                                   | Optimized By AI                                |

# 🔹 **Step 3: Apply Moving Average for Trend Smoothing**

📊 **Goal:** Reduce noise and highlight **true demand patterns** using a **moving average**.

---

## ✅ Overview

Real-world demand data can be **messy**, affected by:

* Sudden sales
* Weather
* Random events

To **smooth out short-term fluctuations**, we use **Moving Averages**.

---

### 📘 What is a Moving Average?

A **moving average** replaces each point in the time series with the **average of its previous N values**.

**Example:**

* 3-Day Moving Average on days 1, 2, 3 = `(D1 + D2 + D3)/3`
* This gives a smoothed curve showing general **trends** 📈

---

## 💡 Benefits:

| Feature                | Why It Helps                               |
| ---------------------- | ------------------------------------------ |
| 📉 Removes Noise       | Reduces random day-to-day changes          |
| 🔄 Highlights Patterns | Makes seasonality/trends visible           |
| 🤖 Useful for ML       | Creates additional features for prediction |

---

## 💻 Python Code: Apply 7-Day Moving Average

```python
# Step 3: Apply Moving Average for Smoothing
```

````python
# Step 4: Train a Basic ML Model 🤖

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Reset index to work with rows
df_reset = df.reset_index(drop=True)

# Feature Engineering
df_reset['day'] = df_reset['date'].dt.day
df_reset['month'] = df_reset['date'].dt.month
df_reset['prev_demand'] = df_reset['demand'].shift(1)

# Drop first row (NaN in prev_demand)
df_reset.dropna(inplace=True)

# ✅ Reset index again after dropping NaNs
df_reset = df_reset.reset_index(drop=True)

# Select features and target
X = df_reset[['day', 'month', 'prev_demand']]
y = df_reset['demand']

# Print to verify
print(X)
print("=========")
print(y)
````

```python
import datetime
```

# 🔹 **Step 6: Inventory Optimization using EOQ & ROP**

📦 **Goal:** Use the forecasted demand to calculate the **Economic Order Quantity (EOQ)** and the **Reorder Point (ROP)** — two fundamental concepts in smart inventory management.

---

## ✅ Overview

Once we have predicted future demand (Step 5), we can now:

* Use it to estimate **how much inventory to order** (EOQ)
* Determine **when to place the order** (ROP)

These help:

* Avoid overstocking and understocking
* Optimize warehouse space and cash flow
* Automate restocking with smart systems

---

## 📘 Concepts Recap:

### 📦 **EOQ (Economic Order Quantity)**

$$
EOQ = \sqrt{\frac{2DS}{H}}
$$

Where:

* $D$ = Forecasted **Annual Demand** (units)
* $S$ = **Ordering cost** per order (Rs or \$)
* $H$ = **Holding cost** per unit per year

---

### 🔔 **ROP (Reorder Point)**

$$
ROP = (\text{Average Daily Demand} \times \text{Lead Time}) + \text{Safety Stock}
$$

---

## 💡 Assumptions for This Lab:

* $S = 100$: Cost to place one order
* $H = 2$: Cost to hold one unit for a year
* **Lead time = 5 days**
* **Safety stock = 50 units**
* Use predicted demand from the **next 7 days**

---

## 💻 Python Code — EOQ & ROP Calculation

```python
import numpy as np

# Step 6: Inventory Optimization 🧮

# Use forecasted 7-day demand to estimate annual demand
daily_avg = forecast_df['predicted_demand'].mean()
annual_demand = daily_avg * 365  # Approximate yearly demand

# Set constants
S = 100  # ordering cost per order
H = 2    # holding cost per unit per year
lead_time_days = 5
safety_stock = 50

# EOQ calculation
EOQ = np.sqrt((2 * annual_demand * S) / H)

# ROP calculation
ROP = (daily_avg * lead_time_days) + safety_stock

# Show results
print("📦 Inventory Optimization Results:")
print(f"✅ Estimated Daily Demand: {daily_avg:.2f} units")
print(f"📐 Economic Order Quantity (EOQ): {EOQ:.2f} units")
print(f"🔔 Reorder Point (ROP): {ROP:.2f} units")
```

## 🧠 What This Means:

* 📐 **EOQ \~ 960**: Place orders in batches of around 960 units for cost efficiency.
* 🔔 **ROP \~ 1054**: Reorder when inventory reaches 1054 units (based on forecast + lead time + buffer).

This allows:

* 📦 Efficient warehouse usage
* 💰 Cost savings
* 🔄 Automated restocking via AI + business rules

````python
%%writefile app.py
# app.py
# ----------------------------
# AI-based Demand Forecasting Dashboard (finalized)
# ----------------------------
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from io import StringIO

# ---------- Streamlit Page Setup ----------
st.set_page_config(page_title="AI Demand Forecasting", layout="wide")
st.title("📦 AI-based Demand Forecasting Dashboard")
st.markdown(
    "Upload your CSV with at least: `date`, `product`, `demand`."
)

# Make uploader obvious in dark mode/wide layout
uploaded_file = st.file_uploader(
    "📁 Upload historical demand CSV",
    type=["csv"],
    label_visibility="visible",
    accept_multiple_files=False,
    help="CSV must contain columns: date, product, demand"
)

# Provide a tiny sample CSV to test quickly
with st.expander("Need a sample CSV? Click to copy/paste"):
    st.code(
        "date,product,demand\n"
        "2024-01-01,A,120\n2024-01-02,A,130\n2024-01-03,A,110\n"
        "2024-01-04,A,140\n2024-01-05,A,115\n2024-01-06,A,150\n"
        "2024-01-07,A,160\n2024-01-08,A,155\n2024-01-09,A,170\n"
        "2024-01-10,A,165\n",
        language="csv"
    )

# ---------- Early exit if no file ----------
if uploaded_file is None:
    st.stop()

# ---------- Load & Validate ----------
@st.cache_data(show_spinner=True)
def load_clean_csv(file) -> pd.DataFrame:
    df = pd.read_csv(file)
    # Standardize column names
    df.columns = [c.strip().lower() for c in df.columns]
    required = {"date", "product", "demand"}
    missing = required - set(df.columns)
    if missing:
        raise ValueError(f"Missing required column(s): {', '.join(sorted(missing))}")

    # Parse date and clean
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df = df.dropna(subset=["date", "demand"])
    # Coerce demand to numeric
    df["demand"] = pd.to_numeric(df["demand"], errors="coerce")
    df = df.dropna(subset=["demand"])

    # Sort
    df = df.sort_values(["product", "date"]).reset_index(drop=True)

    return df

try:
    df = load_clean_csv(uploaded_file)
except Exception as e:
    st.error(f"❌ Could not read the file: {e}")
    st.stop()

# Optional daily aggregation (if duplicates per product-date)
if df.groupby(["product", "date"]).size().max() > 1:
    df = (
        df.groupby(["product", "date"], as_index=False)["demand"]
        .sum()
        .sort_values(["product", "date"])
    )

# ---------- Product picker ----------
products = df["product"].astype(str).unique().tolist()
if len(products) > 1:
    selected_product = st.selectbox("🛍️ Select product", products, index=0)
else:
    selected_product = products[0]
dfp = df[df["product"].astype(str) == str(selected_product)].copy()

# Guard for minimum rows
if len(dfp) < 10:
    st.warning("⚠️ Not enough rows for meaningful modeling. Please upload more history (≥ 10 rows).")
    st.dataframe(dfp.head())
    st.stop()

# ---------- Preview ----------
st.subheader("🔍 Data Preview")
st.dataframe(dfp.head())

# ---------- Demand Over Time ----------
st.subheader("📊 Demand Over Time")
fig = plt.figure(figsize=(10, 4))
sns.lineplot(data=dfp, x="date", y="demand")
plt.xticks(rotation=45)
plt.tight_layout()
st.pyplot(fig)
plt.close(fig)

# ---------- Moving Average ----------
st.subheader("📈 Moving Average (Trend Smoothing)")
ma_col1, ma_col2 = st.columns([1, 3])
with ma_col1:
    window = st.slider("Moving average window (days)", 2, 60, 7, help="Choose smoothing window")
dfp["moving_avg"] = dfp["demand"].rolling(window=window).mean()
fig = plt.figure(figsize=(10, 4))
sns.lineplot(data=dfp, x="date", y="demand", label="Actual")
sns.lineplot(data=dfp, x="date", y="moving_avg", label=f"{window}-Day MA")
plt.xticks(rotation=45)
plt.tight_layout()
st.pyplot(fig)
plt.close(fig)

# ---------- Feature Engineering ----------
st.subheader("🧠 Train ML Model for Forecasting")
dfp["day"] = dfp["date"].dt.day
dfp["month"] = dfp["date"].dt.month
dfp["prev_demand"] = dfp["demand"].shift(1)
dfp = dfp.dropna().reset_index(drop=True)

X = dfp[["day", "month", "prev_demand"]]
y = dfp["demand"]

# Split chronologically (no shuffle)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=min(0.2, 1 - 1/len(dfp)), shuffle=False
)

model_choice = st.selectbox("Choose a model", ["Linear Regression", "Random Forest"])
if model_choice == "Linear Regression":
    model = LinearRegression()
else:
    model = RandomForestRegressor(n_estimators=200, random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

met_col1, met_col2 = st.columns(2)
with met_col1:
    st.metric("MAE", f"{mae:.2f}")
with met_col2:
    st.metric("RMSE", f"{rmse:.2f}")

# Plot predictions vs actual
fig = plt.figure(figsize=(10, 4))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(y_pred, label="Predicted", marker="x")
plt.title("📈 Forecasting Results (Holdout)")
plt.legend()
plt.tight_layout()
st.pyplot(fig)
plt.close(fig)

# ---------- Future Forecast (iterative) ----------
st.subheader("🔮 Future Forecast")
future_days = st.slider("Forecast horizon (days)", 1, 60, 14)

last_date = dfp["date"].max()
future_dates = pd.date_range(last_date + pd.Timedelta(days=1), periods=future_days, freq="D")

# Build features iteratively using prev_demand as last known/predicted
prev_val = dfp["demand"].iloc[-1]
future_preds = []
for d in future_dates:
    features = pd.DataFrame(
        [[d.day, d.month, prev_val]],
        columns=["day", "month", "prev_demand"]
    )
    pred = float(model.predict(features)[0])
    future_preds.append(pred)
    prev_val = pred

future_df = pd.DataFrame({"date": future_dates, "forecast": future_preds})
fig = plt.figure(figsize=(10, 4))
sns.lineplot(x=dfp["date"], y=dfp["demand"], label="History")
sns.lineplot(x=future_df["date"], y=future_df["forecast"], label="Forecast")
plt.xticks(rotation=45)
plt.tight_layout()
st.pyplot(fig)
plt.close(fig)

# ---------- Inventory Optimization (EOQ + ROP) ----------
st.subheader("📦 Inventory Optimization (EOQ + ROP)")

# Estimate daily average demand for the chosen product
# If dates are not consecutive, use overall mean as "daily average" proxy
daily_avg = dfp["demand"].mean()

ic1, ic2, ic3 = st.columns(3)
with ic1:
    order_cost = st.number_input("Ordering cost per order", value=100.0, min_value=0.0, step=10.0)
with ic2:
    holding_cost = st.number_input("Holding cost per unit per period", value=5.0, min_value=0.0001, step=0.5)
with ic3:
    lead_time = st.number_input("Lead time (days)", value=7, min_value=0, step=1)

eoq = np.sqrt((2 * daily_avg * order_cost) / holding_cost) if holding_cost > 0 else np.nan
reorder_point = daily_avg * lead_time

st.markdown(f"- **📦 EOQ**: `{eoq:.2f}` units")
st.markdown(f"- **⏰ Reorder Point (ROP)**: `{reorder_point:.2f}` units")

# Visualize EOQ/ROP vs history
fig = plt.figure(figsize=(10, 4))
sns.lineplot(data=dfp, x="date", y="demand", label="Demand")
plt.axhline(eoq, linestyle="--", label=f"EOQ ({eoq:.2f})")
plt.axhline(reorder_point, linestyle="--", label=f"ROP ({reorder_point:.2f})")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
st.pyplot(fig)
plt.close(fig)

# ---------- Downloads ----------
st.subheader("⬇️ Download Results")

# Holdout predictions (test set) with dates aligned
test_index = X_test.index
results_df = pd.DataFrame({
    "date": dfp.loc[test_index, "date"].values,
    "product": selected_product,
    "actual": y_test.values,
    "predicted": y_pred,
})
st.download_button(
    "📥 Download Holdout Predictions (CSV)",
    data=results_df.to_csv(index=False),
    file_name="holdout_predictions.csv",
    mime="text/csv",
)

# Future forecast download
future_out = future_df.copy()
future_out.insert(1, "product", selected_product)
st.download_button(
    "📥 Download Future Forecast (CSV)",
    data=future_out.to_csv(index=False),
    file_name="future_forecast.csv",
    mime="text/csv",
)

st.success("✅ Done. Adjust inputs to re-run modeling and inventory suggestions.")
````
